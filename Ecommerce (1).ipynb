#### **Cleaning Missing Value**

1. Pembersihan pada `order_reviews`



```python
# Mengisi missing values di order_reviews
order_reviews['review_comment_title'] = order_reviews['review_comment_title'].fillna('')
order_reviews['review_comment_message'] = order_reviews['review_comment_message'].fillna('')
```

```python
print("Missing values in orders_clean:")
print(order_reviews.isnull().sum())
```

2. Pembersihan pada `orders`





```python
# Menghapus baris dengan missing values di order_approved_at
orders = orders.dropna(subset=['order_approved_at'])

# Mengisi missing values dengan modus
orders['order_delivered_carrier_date'] = orders['order_delivered_carrier_date'].fillna(orders['order_delivered_carrier_date'].mode()[0])
orders['order_delivered_customer_date'] = orders['order_delivered_customer_date'].fillna(orders['order_delivered_customer_date'].mode()[0])
```

```python
print("Missing values in orders_clean:")
print(orders.isnull().sum())
```

3. Pembersihan pada `products`


```python
# Menghapus baris yang memiliki missing values di product_category_name
products= products.dropna(subset=['product_category_name'])

# Mengisi kolom dengan mean
products['product_name_lenght'] = products['product_name_lenght'].fillna(products['product_name_lenght'].mean())
products['product_description_lenght'] = products['product_description_lenght'].fillna(products['product_description_lenght'].mean())
products['product_photos_qty'] = products['product_photos_qty'].fillna(products['product_photos_qty'].mean())

# Mengisi kolom numerik lainnya
products['product_weight_g'] = pd.to_numeric(products['product_weight_g'], errors='coerce')
products['product_length_cm'] = pd.to_numeric(products['product_length_cm'], errors='coerce')
products['product_height_cm'] = pd.to_numeric(products['product_height_cm'], errors='coerce')
products['product_width_cm'] = pd.to_numeric(products['product_width_cm'], errors='coerce')

products[['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']] = \
products[['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']].fillna(products[['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']].mean())
```

```python
print("Missing values in orders_clean:")
print(products.isnull().sum())
```

* Mengidentifikasi masalah logistik: Jika banyak pengiriman yang terlambat, ini bisa menjadi indikasi adanya masalah pada proses logistik, seperti kurangnya kapasitas gudang, masalah transportasi, atau kesalahan dalam perencanaan rute.

* Meningkatkan efisiensi operasi: Dengan mengidentifikasi akar penyebab keterlambatan, perusahaan dapat mengambil langkah-langkah untuk meningkatkan efisiensi operasi dan mengurangi biaya.



Process

1. Merge Datasets (Orders and Products): merge the orders_cleaned with the order_items_cleaned dataset to get product details.

2. Create Delivery Time Metrics: Calculate actual delivery time and compare it with estimated delivery time.

3. Visualize Delivery Performance: A bar chart to visualize the percentage of orders delivered on time vs. late.


```python
# delivery time
merged_data['actual_delivery_time'] = (merged_data['order_delivered_customer_date'] - merged_data['order_purchase_timestamp']).dt.days
merged_data['estimated_delivery_time'] = (merged_data['order_estimated_delivery_date'] - merged_data['order_purchase_timestamp']).dt.days

# check if orders are delivered earlier or late
merged_data['delivery_status'] = np.where(merged_data['actual_delivery_time'] <= merged_data['estimated_delivery_time'], 'On time', 'Late')
```

```python
print(merged_data['delivery_status'].value_counts())
```

```python
# Plot delivery performance
plt.figure(figsize=(8, 5))
sns.countplot(data=merged_data, x='delivery_status')
plt.title('Order Delivery Performance')
plt.show()
```

### Q5 : Bagaimana waktu persetujuan pesanan mempengaruhi jadwal pengiriman?

Pertanyaan ini bertujuan untuk memahami hubungan antara proses internal dan kinerja pengiriman. Dengan mengetahui bagaimana waktu persetujuan pesanan mempengaruhi waktu pengiriman, bisnis dapat:

* Mengidentifikasi bottleneck: Mengidentifikasi tahap-tahap dalam proses pemesanan yang menyebabkan keterlambatan.

* Mengoptimalkan proses bisnis: Mengoptimalkan proses bisnis untuk mengurangi waktu tunggu dan meningkatkan efisiensi.

* Meningkatkan prediksi: Membuat prediksi yang lebih akurat tentang waktu pengiriman berdasarkan waktu persetujuan pesanan.

Process:

1. Calculate Delivery Delay and Approval Time

2. Calculate Correlation Between Approval Time and Delivery Delay

3. Create a Scatter Plot to Visualize the Correlation



```python
# delivery delay

orders_cleaned['delay'] = (orders_cleaned['order_delivered_customer_date'] - orders_cleaned['order_estimated_delivery_date']).dt.days

# time from approval to delivery

orders_cleaned['approval_time'] = (orders_cleaned['order_delivered_customer_date'] - orders_cleaned['order_approved_at']).dt.days

# correlation between delays and approval times

print(orders_cleaned[['approval_time', 'delay']].corr())
```

0.992113 is very close to 1, indicating a very strong positive correlation between approval_time and delay. This means that as the approval time increases, the delivery delay also tends to increase. In other words, the longer it takes to approve an order, the more likely it is that the delivery will be delayed.

```python
# scatter plot

sns.scatterplot(x='approval_time', y='delay', data=orders_cleaned)

plt.title('Correlation between Approval Time and Delivery Delay')
plt.xlabel('Approval Time (days)')
plt.ylabel('Delay (days)')

# plot
plt.show()
```

### RFM Analysis



*   Bagaimana distribusi nilai keterkinian pelanggan?

*   Berapa frekuensi pembelian di antara berbagai segmen pelanggan?

*   Bagaimana distribusi nilai moneter di antara pelanggan?



```python
# Calculating Recency, Frequency, and Monetary

current_date = merged_data['order_purchase_timestamp'].max()

rfm_data = merged_data.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (current_date - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency
    'price': 'sum'        # Monetary
}).reset_index()

rfm_data.columns = ['customer_id', 'Recency', 'Frequency', 'Monetary']
```

```python
# View 10 customers with the highest Monetary,recency, frequency value
print(rfm_data.sort_values(by='Monetary', ascending=False).head(10),'\n')
print(rfm_data.sort_values(by='Recency', ascending=False).head(10),'\n')
print(rfm_data.sort_values(by='Monetary', ascending=False).head(10),'\n')
```

```python
# RFM histogram
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Plot Distribusi Recency
sns.histplot(rfm_data['Recency'], bins=20, ax=ax[0], color='blue')
ax[0].set_title('Distribusi Recency', fontsize=16)
ax[0].set_xlabel('Recency (days)', fontsize=14)

# Plot Distribusi Frequency
sns.histplot(rfm_data['Frequency'], bins=20, ax=ax[1], color='green')
ax[1].set_title('Distribusi Frequency', fontsize=16)
ax[1].set_xlabel('Frequency', fontsize=14)

# Plot Distribusi Monetary
sns.histplot(rfm_data['Monetary'], bins=20, ax=ax[2], color='red')
ax[2].set_title('Distribusi Monetary', fontsize=16)
ax[2].set_xlabel('Monetary (Revenue)', fontsize=14)

plt.tight_layout()
plt.show()
```
